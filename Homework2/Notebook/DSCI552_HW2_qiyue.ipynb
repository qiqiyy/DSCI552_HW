#### (i) KNN Regression:
i. Perform k-nearest neighbor regression for this dataset using both normalized and raw features. Find the value of k âˆˆ {1, 2, . . . , 100} that gives you the best fit. Plot the train and test errors in terms of 1/k.

```python
# Model 1 for raw features
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
X_base = data.loc[:, ['AT', 'V', 'AP', 'RH']]
y = data.loc[:, ['PE']]
X_train, X_test, y_train, y_test = train_test_split(X_base, y, test_size = 0.3, random_state = 1)

test_error = []
train_error = []
N = []
r = range(1, 101)
for i in r:
    k = i
    N.append(1/k)
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)
    # Calculate test error rate
    y_preds1 = knn.predict(X_test)
    test_error.append(mean_squared_error(y_test, y_preds1))
    # Calculate train error rate
    y_preds2 = knn.predict(X_train)
    train_error.append(mean_squared_error(y_train, y_preds2))

# Plot train error rates & test error rate on the map   
plt.figure(figsize=(10, 7))
plt.plot(N,train_error, color='red',label='Training error rate')
plt.plot(N,test_error, color='blue',label='Testing error rate')
plt.xlabel('1/K')
plt.ylabel('Error')
plt.legend()
```

```python
# Find out the optimal k value 
k = list(r)[test_error.index(min(test_error))]
print("Minimum error using raw features : ", min(test_error), "at K =", k)
```

```python
# Model 2 for normalized features
from sklearn.preprocessing import MinMaxScaler
# fit scaler on training data
norm = MinMaxScaler().fit(X_train)
# transform training data
X_train_norm = norm.transform(X_train)
# transform testing dataabs
X_test_norm = norm.transform(X_test)

test_error = []
train_error = []
N = []
r = range(1, 101)
for i in r:
    k = i
    N.append(1/k)
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train_norm, y_train)
    # Calculate test error rate
    y_preds1 = knn.predict(X_test_norm)
    test_error.append(mean_squared_error(y_test, y_preds1))
    # Calculate train error rate
    y_preds2 = knn.predict(X_train_norm)
    train_error.append(mean_squared_error(y_train, y_preds2))

# Plot train error rates & test error rate on the map   
plt.figure(figsize=(10, 7))
plt.plot(N,train_error, color='red',label='Training error rate')
plt.plot(N,test_error, color='blue',label='Testing error rate')
plt.xlabel('1/K')
plt.ylabel('Error')
plt.legend()
```

```python
# Find out the optimal k value 
k = list(r)[test_error.index(min(test_error))]
print("Minimum error using normalized features: ", min(test_error), "at K =", k)
```

#### (a) Compute the Euclidean distance between each observation and the test point, X1 = X2 = X3 = 0. 

```python
from scipy.spatial import distance
data = [[0,3,0,'Red'],[2,0,0,'Red'],[0,1,3,'Red'],[0,1,2,'Green'],[-1,0,1,'Green'],[1,1,1,'Red']]

Dis = []
for i in range(6):
    X = data[i][:3]
    Y = [0, 0, 0]
    dist = distance.euclidean(X, Y)
    Dis.append(dist)

data = pd.DataFrame(data, index=[1,2,3,4,5,6], columns=['X1','X2','X3','Y'])
data['Dis'] = Dis
data
```
